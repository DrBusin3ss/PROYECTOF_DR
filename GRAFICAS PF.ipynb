```python
#GRAFICO 1

import matplotlib.pyplot as plt

# Datos
categories = ['Gen Z', 'Gen X', 'millennials', 'babyboomer']
count = [1, 2, 3, 4]

# Crear el gráfico de barras
plt.bar(categories, count)

# Agregar etiquetas de datos
for i, v in enumerate(count):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Personalizar el gráfico
plt.xlabel('Generación')
plt.ylabel('Cantidad')
plt.title('Cantidad de clientes por generación')

# Mostrar el gráfico
plt.show()
```

```python
#2
#Grafico de Pastel

import matplotlib.pyplot as plt

# Datos
labels = ['Male', 'Female']
sizes = [1, 2]

# Crear el gráfico de pastel
plt.pie(sizes, labels=labels)

# Agregar etiquetas de datos
total = sum(sizes)
for i, v in enumerate(sizes):
    percentage = v / total * 100
    plt.text(0, -0.8, f'{labels[i]}: {v} ({percentage:.1f}%)', ha='center', va='center')
# Agregar una leyenda a la derecha de cada porción
plt.legend(title="Genero", loc="center left", bbox_to_anchor=(1, 0.5))

# Mostrar el gráfico
plt.show()
```

```python
#3
# GRAFICO DE DISPERCION 
import matplotlib.pyplot as plt

# Datos
x = [22, 41, 30, 35, 50]  # Edades
y = [598, 584, 757, 232, 533]  # Gasto total en USD

# Crear el gráfico de dispersión
plt.scatter(x, y)

# Agregar etiquetas de datos
for i, txt in enumerate(y):
    plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Personalizar el gráfico
plt.xlabel('Edad')
plt.ylabel('Gasto total USD')
plt.title('Gasto total en función de la edad')

# Mostrar el gráfico
plt.show()
```
